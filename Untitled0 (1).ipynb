7. What is the difference between “==” and “is” operators in Python

->The difference between == and is in Python is a common question, and it’s important to understand value vs identity.

1. == operator (Equality)

Checks whether the values of two objects are the same.

It does not care if they are the same object in memory.

a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True, because their values are the same

2. is operator (Identity)

Checks whether the two objects are actually the same object in memory.

It’s about object identity, not value.

a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)  # False, because a and b are different objects in memory

c = a
print(a is c)  # True, because c points to the same object as a

Quick Table for Memory
Operator	Checks for	Example Result
==	Value equality	[1,2]==[1,2] → True
is	Object identity	[1,2] is [1,2] → False

9.What is type casting in Python


->Type casting in Python is the process of converting a variable from one data type to another.

Why do we need type casting?

To perform operations between different data types.

To change input data types (e.g., user input is always a string).

To match the expected data type in a function or calculation.

Types of Type Casting in Python

Implicit Type Casting (Type Conversion done automatically)

Python automatically converts one data type to another without user intervention.

Usually happens with numeric types.

x = 5        # int
y = 2.5      # float
z = x + y    # int + float → float automatically
print(z)     # Output: 7.5
print(type(z))  # <class 'float'>


Explicit Type Casting (Type Conversion done manually)

The programmer manually converts a variable using functions like int(), float(), str(), list(), etc.

x = "10"        # str
y = int(x)      # convert string to integer
print(y + 5)    # Output: 15
print(type(y))  # <class 'int'>

Common Type Casting Functions:
Function	Converts To
int()	Integer
float()	Float
str()	String
list()	List
tuple()	Tuple
set()	Set

1. Implicit Type Casting (Type Conversion done automatically)

Python automatically converts one data type to another.

12.How does the elif statement work

->The elif statement in Python stands for “else if”. It is used to check multiple conditions after an initial if statement.

How elif works:

Python first checks the if condition.

If the if condition is True, its block executes, and Python skips all other checks.

If the if condition is False, Python moves to the elif condition(s).

The first elif condition that is True will execute, and Python skips the remaining elif or else blocks.

If none of the if or elif conditions are True, the else block (if present) executes.

Example:
x = 15

if x > 20:
    print("x is greater than 20")
elif x > 10:
    print("x is greater than 10")  # ✅ This will run
elif x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")


Output:

x is greater than 10


Notice that even though x > 5 is True, Python does not check further after the first True condition (x > 10).

13. What is the difference between for and while loops


->The difference between for and while loops in Python comes down to how they repeat code and what they check.

1. for loop

Used when you know in advance how many times you want to repeat something.

Iterates over a sequence (like list, tuple, string, or range).

Example:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)


Output:

apple
banana
cherry

2. while loop

Used when you want to repeat as long as a condition is True.

You may not know in advance how many times it will run.

Example:

count = 1
while count <= 3:
    print(count)
    count += 1


Output:

1
2
3

Quick Comparison Table
Feature	for loop	while loop
Use case	Known number of iterations	Unknown number, depends on condition
Iterates over	Sequence (list, tuple, string, range)	Condition (True/False)
Syntax	for item in sequence:	while condition:
Risk of infinite loop	Low	High (if condition never becomes False)

14.Describe a scenario where a while loop is more suitable than a for loop.


->A while loop is more suitable than a for loop when the number of iterations is not known in advance and depends on a condition being True.

Scenario Example: User Login Attempts

Imagine you are creating a login system where the user can try to enter a password until it’s correct. You don’t know in advance how many attempts the user will need.

correct_password = "Python123"
attempt = input("Enter password: ")

while attempt != correct_password:
    print("Incorrect password. Try again.")
    attempt = input("Enter password: ")

print("Login successful!")


Why while is better here:

You don’t know how many attempts the user will make.

The loop continues as long as the condition (attempt != correct_password) is True.

A for loop would require a fixed number of iterations, which doesn’t fit this scenario.

Other examples where while loops are better:

Repeating a menu until the user chooses “exit”.

Reading data from a file until the end of the file.

Waiting for a sensor value to reach a threshold in real-time applications.

```python
#6.Write a function to demonstrate the behavior of mutable and immutable arguments.

# Function to modify arguments
def modify(mutable_arg, immutable_arg):
    # Modify the mutable argument (list)
    mutable_arg.append(100)
    # Try to modify the immutable argument (integer)
    immutable_arg += 100
    print("Inside function - mutable_arg:", mutable_arg)
    print("Inside function - immutable_arg:", immutable_arg)

# Mutable object: list
my_list = [1, 2, 3]

# Immutable object: integer
my_num = 10

print("Before function call:")
print("my_list:", my_list)
print("my_num:", my_num)

# Call the function
modify(my_list, my_num)

print("\nAfter function call:")
print("my_list:", my_list)  # List is changed
print("my_num:", my_num)    # Integer is unchanged
```

```python
#7.Write a function to demonstrate the behavior of mutable and immutable arguments.


def modify_values(num, lst):
    """
    Demonstrates behavior of immutable (num) and mutable (lst) arguments.
    """
    print("Inside function before modification:")
    print("num =", num)
    print("lst =", lst)

    # Attempt to modify the immutable argument
    num += 10

    # Modify the mutable argument
    lst.append(100)

    print("\nInside function after modification:")
    print("num =", num)
    print("lst =", lst)

# Driver code
x = 5            # Immutable (integer)
my_list = [1, 2, 3]  # Mutable (list)

print("Before function call:")
print("x =", x)
print("my_list =", my_list)

modify_values(x, my_list)

print("\nAfter function call:")
print("x =", x)          # x remains unchanged
print("my_list =", my_list)  # my_list is changed
```

```python
#9.Write a Python program to convert user input from string to integer, float, and boolean types.


# Program to convert user input to different types

# Taking input from the user
user_input = input("Enter something: ")

# Converting to integer
try:
    int_value = int(user_input)
except ValueError:
    int_value = "Cannot convert to integer"

# Converting to float
try:
    float_value = float(user_input)
except ValueError:
    float_value = "Cannot convert to float"

# Converting to boolean
# Any non-empty string is considered True, empty string is False
bool_value = bool(user_input)

# Displaying results
print("\nConversion results:")
print(f"Original input (string): {user_input}")
print(f"As integer: {int_value}")
print(f"As float: {float_value}")
print(f"As boolean: {bool_value}")
```

```python
#10.Write code to demonstrate type casting with list elements.

# Program to demonstrate type casting with list elements

# Original list with mixed types (strings and numbers)
original_list = ["10", "20", "30", "40.5", "50"]

print("Original list:", original_list)

# Converting all elements to integers (where possible)
int_list = []
for item in original_list:
    try:
        int_list.append(int(float(item)))  # first convert to float to handle "40.5", then to int
    except ValueError:
        int_list.append(item)  # if conversion fails, keep original

print("List converted to integers:", int_list)

# Converting all elements to floats
float_list = []
for item in original_list:
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(item)

print("List converted to floats:", float_list)

# Converting all elements to strings
str_list = [str(item) for item in original_list]
print("List converted to strings:", str_list)

# Converting all elements to booleans
bool_list = [bool(item) for item in original_list]  # non-empty strings → True
print("List converted to booleans:", bool_list)
```

```python
 #13.Write a Python program to find the sum of all even numbers between 1 and 50

 # Program to find the sum of all even numbers between 1 and 50

# Initialize sum
sum_even = 0

# Loop through numbers from 1 to 50
for num in range(1, 51):
    if num % 2 == 0:  # Check if the number is even
        sum_even += num

# Display the result
print("The sum of all even numbers between 1 and 50 is:", sum_even)
```

```python
#14.Write a program to reverse a string using a while loop.

# Program to reverse a string using a while loop

# Taking input from the user
original_string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(original_string) - 1  # Start from the last character

# Using while loop to reverse the string
while index >= 0:
    reversed_string += original_string[index]
    index -= 1  # Move to the previous character

# Display the reversed string
print("Reversed string:", reversed_string)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

# Program to calculate factorial using a while loop

# Taking input from the user
num = int(input("Enter a non-negative integer: "))

# Initialize factorial result
factorial = 1
counter = num  # start counting down from num

# Using while loop to calculate factorial
while counter > 0:
    factorial *= counter
    counter -= 1  # move to the next smaller number

# Display the result
print(f"The factorial of {num} is {factorial}")
```
